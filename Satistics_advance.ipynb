#Que 1 What is a random variable in probability theory
A **random variable** in probability theory is a numerical outcome of a random experiment or process. It assigns a real number to each outcome in a sample space.

There are two main types of random variables:

1. **Discrete Random Variable**: Takes on a finite or countable number of values (e.g., rolling a die).
2. **Continuous Random Variable**: Takes on an infinite number of possible values, usually within a given range (e.g., height, weight).

A random variable is often denoted by a letter like **X**, and its values are governed by a probability distribution.

#Que 2 What are the types of random variables
There are two main types of random variables:

1. **Discrete Random Variable**:

   * Takes a finite or countable number of distinct values.
   * Examples: Number of heads in 10 coin flips, the roll of a die.

2. **Continuous Random Variable**:

   * Takes an infinite number of possible values, typically within a given range.
   * Examples: Height, weight, time, temperature.

#Que 3  What is the difference between discrete and continuous distributions

* **Discrete Distribution**: Deals with random variables that can take on a finite or countable number of distinct values. Probabilities are assigned to individual outcomes.

  * Example: **Binomial Distribution** (number of heads in a series of coin flips).

* **Continuous Distribution**: Deals with random variables that can take on an infinite number of values within a given range. Probabilities are assigned to intervals rather than specific outcomes.

  * Example: **Normal Distribution** (height, weight, or time).

The main difference is that discrete distributions involve countable outcomes, while continuous distributions involve uncountable, real-valued outcomes.

#Que 4 What are probability distribution functions (PDF)
A PDF gives the likelihood of a value or range of values for a random variable; it’s PMF for discrete and a density curve for continuous.

#Que 5 How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)
A CDF gives the probability that a variable is less than or equal to a value; a PDF gives the likelihood of exact values (discrete) or ranges (continuous).

#que 6 = What is a discrete uniform distribution
It’s a distribution where all outcomes have equal probability, like rolling a fair die (each side has 1/n chance).

#Que 7 What are the key properties of a Bernoulli distribution
Two outcomes (0 or 1), one trial, probability of success $p$, and expected value $p$, variance $p(1 - p)$.

#Que 8 What is the binomial distribution, and how is it used in probability
It models the number of successes in $n$ independent Bernoulli trials with probability $p$; used for repeated yes/no experiments.

#Que 9 = What is the Poisson distribution and where is it applied
It models the number of events in a fixed interval of time or space, given a constant average rate; used in queues, arrivals, or rare event counts.

#Que 10 What is a continuous uniform distribution
It’s a distribution where all values in an interval $[a, b]$ are equally likely; PDF is constant over the interval.

#Que 12 What is the standard normal distribution, and why is it important
It’s a normal distribution with mean 0 and standard deviation 1; used to compute probabilities and z-scores for any normal variable.

#Que 13 What is the Central Limit Theorem (CLT), and why is it critical in statistics
The CLT states that the distribution of the sample mean approaches normality as sample size increases, regardless of the population's distribution. It’s critical because it allows for inferential statistics using normal distribution approximations.

#Que 14How does the Central Limit Theorem relate to the normal distribution
The CLT shows that sample means from any distribution will approximate a normal distribution as the sample size grows, even if the original data is not normal.

#Que 15 What is the application of Z statistics in hypothesis testing
Z-statistics are used to test hypotheses about population parameters when the population is normally distributed, helping determine if a sample mean significantly differs from the population mean.

#Que 16 = How do you calculate a Z-score, and what does it represent
The Z-score is calculated as:

$$
Z = \frac{X - \mu}{\sigma}
$$

Where $X$ is the value, $\mu$ is the mean, and $\sigma$ is the standard deviation. It represents how many standard deviations a value is from the mean.

#Que 17 What are point estimates and interval estimates in statistics
* **Point estimate**: A single value used to estimate a population parameter (e.g., sample mean).
* **Interval estimate**: A range of values (e.g., confidence interval) that likely contains the population parameter.

#Que 18 What is the significance of confidence intervals in statistical analysis
Confidence intervals provide a range of values within which a population parameter is likely to fall, offering an estimate of precision and uncertainty in statistical analysis.

#Que 19 What is the relationship between a Z-score and a confidence interval
A Z-score represents the number of standard deviations a data point is from the mean, and it’s used to calculate the confidence interval, which provides a range of values around the sample mean for a given confidence level.

#Que 20  How are Z-scores used to compare different distributions
Z-scores standardize values from different distributions, allowing comparison by converting them to a common scale (standard normal distribution).

#Que 21 What are the assumptions for applying the Central Limit Theorem
The assumptions for applying the Central Limit Theorem are:

1. The sample size is sufficiently large (usually $n \geq 30$).
2. The samples are independent.
3. The population has a finite mean and variance.

#Que 22 What is the concept of expected value in a probability distribution
The expected value is the long-term average or mean of a random variable, calculated as the sum of each possible outcome weighted by its probability.

#Que 23 How does a probability distribution relate to the expected outcome of a random variable
A probability distribution shows the likelihood of each possible outcome, and the expected value is the weighted average of those outcomes, representing the "center" or most likely average result over many trials.
